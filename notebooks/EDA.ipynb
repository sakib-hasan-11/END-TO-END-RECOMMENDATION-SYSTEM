```python
article= pd.read_parquet('sample_data/articles_sample.parquet')
customer= pd.read_parquet('sample_data/customers_sample.parquet')
transaction= pd.read_parquet('sample_data/transactions_sample.parquet')
```

```python
no_need =['article_id', 'product_code']
vals={}
for el in article.columns:
    if el in no_need : 
        continue
    else :
        vals[el]=len(article[el].unique())

```

```python
df=transaction.merge(
    right=customer[['customer_id', 'FN', 'Active', 'club_member_status','fashion_news_frequency', 'age', 'postal_code']],
    on='customer_id'
).groupby('customer_id').agg(
    total_sales=('price','sum'),
    avg_sales=('price','mean'),
    total_orders=('price','count')
).reset_index()
```
